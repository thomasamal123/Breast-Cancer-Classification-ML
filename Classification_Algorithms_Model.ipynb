## IQR Method To Find The Outliers

```python
# Selecting feature columns (exclude target)
features = df.drop('target', axis=1)

# Calculate Q1 and Q3
Q1 = features.quantile(0.25)
Q3 = features.quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1
# Condition for detecting outliers
outlier_condition = ((features < (Q1 - 1.5 * IQR)) | 
                     (features > (Q3 + 1.5 * IQR)))

# Count outliers per column
outliers_per_column = outlier_condition.sum()

outliers_per_column

```

## Train–Test Split and Scale the Training and Testing Data

```python
x = df.drop('target', axis = 1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, random_state=42, test_size=0.2
)
```

```python
from sklearn.preprocessing import StandardScaler

# Create scaler object
scaler = StandardScaler()

# Fit on TRAINING data only
scaler.fit(X_train)

# Transform both sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

```python
# I first checked the dataset for missing values to ensure the model does not get affected by incomplete data.
# Since there were no missing values in this dataset, I did not need to apply any imputation steps.

# I then checked for outliers using the IQR method. Outliers can affect distance-based models like KNN,
# but since the dataset is already clean and balanced, I kept the original data without removing anything.

# I did not apply label encoding because the breast cancer dataset already has numerical features
# and the target variable is provided as 0 and 1, which is suitable for classification models.

# Before training the models, I performed feature scaling using StandardScaler.
# Scaling is important because algorithms like KNN, Logistic Regression, and SVM
# can give wrong results when the features are on different scales.

# To avoid data leakage, I fitted the scaler only on X_train and applied transform on X_test.
# This ensures that information from the test set does not influence the model during training.
```

## Build Classification Models and Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

def evaluate_model(model, X_train, X_test, y_train, y_test):
    """
    This function trains a model and prints evaluation results.
    """
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print("Model:", model.__class__.__name__)
    print("Accuracy:", acc)
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("="*60)

    return acc
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

models = [
    LogisticRegression(max_iter=500),
    DecisionTreeClassifier(),
    SVC(),
    RandomForestClassifier()
]

for m in models:
    evaluate_model(m, X_train, X_test, y_train, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(
    n_neighbors=5,
    algorithm='brute',   
    metric='minkowski', 
    p=2                    # Euclidean distance
)

knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

print("Model: KNeighborsClassifier")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Create a function that returns metrics for each modelnm 
def get_classification_results(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    return {
        "Model Used": model.__class__.__name__,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred)
    }

# Collect results from all models
results = {
    "Logistic Regression": get_classification_results(LogisticRegression(max_iter=500), X_train, X_test, y_train, y_test),
    "Decision Tree": get_classification_results(DecisionTreeClassifier(), X_train, X_test, y_train, y_test),
    "Random Forest": get_classification_results(RandomForestClassifier(), X_train, X_test, y_train, y_test),
    "Support Vector Machine": get_classification_results(SVC(), X_train, X_test, y_train, y_test),
    "KNN Classifier": get_classification_results(KNeighborsClassifier(n_neighbors=5), X_train, X_test, y_train, y_test)
}

# Store results in DataFrame
results_df = pd.DataFrame(results).T
results_df
```

## Working of Classification Models

```python
# Logistic Regression works by applying a sigmoid function to a linear equation, converting values into probabilities between 0 and 1 for classification.
# A Decision Tree repeatedly splits the data based on feature values, forming branches until it classifies samples into final leaf nodes.
# KNN classifies a new sample by checking the 'k' closest data points and assigning the majority class among those neighbors.
# Random Forest is an ensemble method that builds many decision trees on bootstrapped samples (bagging) and averages their predictions.
# SVM finds the best boundary (hyperplane) that separates classes with the maximum margin between them.
```

## Comparing Train and Test Accuracy 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier


def get_classification_results(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred  = model.predict(X_test)

    train_acc = accuracy_score(y_train, train_pred)
    test_acc  = accuracy_score(y_test, test_pred)

    return {
        "Model used": model.__class__.__name__,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc
    }

# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "KNN": KNeighborsClassifier()
}

results = {}

for name, model in models.items():
    results[name] = get_classification_results(model, X_train, X_test, y_train, y_test)

results_df = pd.DataFrame(results).T
results_df
```

## Best Model Selection

```python
# Final Model Selection

# After comparing all the models, Logistic Regression performed the best overall.
# It achieved one of the highest test accuracies (0.956) while keeping the training
# accuracy (0.958) close to the test score. This shows that the model is not
# overfitting and generalizes well to new unseen data.
# Random Forest also gave a high test accuracy, but its training accuracy was 1.0,
# which is a clear sign of overfitting. For this reason, Random Forest was not
# chosen as the final model and same case for Decision Tree as well. 

# Decision Tree and Random Forest both overfitted heavily, SVM performed lower, and although KNN achieved the same test accuracy as Logistic Regression, its lower training accuracy shows it is less stable and generalizes slightly weaker than Logistic Regression. 

# Therefore, the final chosen model is Logistic Regression because it provides the
# best balance between accuracy and generalization.
```
